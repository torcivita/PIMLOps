```python
import numpy as np
import pandas as pd
import os
import os.path
```

Render tiene una restricción de memoria por lo que vamos a reducir los archivos utilizados en la API a un 10%. Esto es simplemente para poder montarlo en Render.

Abrimos los archivos utilizados en el archivo `main.py`

```python
items_games = pd.read_parquet('..\\Archivos API\\items_games_API.parquet')
reviews_games = pd.read_parquet('..\\Archivos API\\reviews_games.parquet')
steam_games = pd.read_parquet('..\\Archivos API\\steam_games_API.parquet')
```

Creamos Dataframes reducidos con el 10% de datos aleatorios

```python
# El valor random_state es para poder replicar la muestra, siempre que tenga el mismo numero, devolverá el mismo Dataframe
items_games_reducido = items_games.sample(frac=0.1, random_state=50)
reviews_games_reducido = reviews_games.sample(frac=0.1, random_state=50)
steam_games_reducido = steam_games.sample(frac=0.1, random_state=50)
```

Reindexamos

```python
items_games_reducido.reset_index(drop=True, inplace=True)
reviews_games_reducido.reset_index(drop=True, inplace=True)
steam_games_reducido.reset_index(drop=True, inplace=True)
```

Comparamos y vemos que efectivamente los Dataframes reducidos contienen un 10% de los datos de los Dataframes originales
